CSV Read

```python
import csv

filename = targetUser +"/_review.csv"
reviews = []

with open(filename, 'r') as f:
    reader = csv.DictReader(f)
    
    for row in reader:
        reviews.append(dict(row))
        
print(reviews)
    
```

```python
filename = targetUser +"/_rating.csv"
ratings = []

with open(filename, 'r') as f:
    reader = csv.DictReader(f)
    
    for row in reader:
        ratings.append(dict(row))
        
print(ratings)
```

Plotting Graph

```python
import gmplot

#Only Reviews
latitudes = []
longitudes = []

if reviews:
    for review in reviews:
        latitudes.append(float(review['latitude']))
        longitudes.append(float(review['longitude']))

    heatmap = gmplot.GoogleMapPlotter(latitudes[0], longitudes[0], 12) 
    heatmap.heatmap(latitudes[1:], longitudes[1:])
    heatmap_file = targetUser + "/_review_heatmap.html"
    heatmap.draw(heatmap_file)
```

```python
latitudes = []
longitudes = []

#Only Ratings
if ratings:
    for rating in ratings:
        latitudes.append(float(rating['latitude']))
        longitudes.append(float(rating['longitude']))

    heatmap = gmplot.GoogleMapPlotter(latitudes[0], longitudes[0], 12) 
    heatmap.heatmap(latitudes[1:], longitudes[1:])
    heatmap_file = targetUser + "/_rating_heatmap.html"
    heatmap.draw(heatmap_file)
    
```

```python
latitudes = {"rev":[], "rate":[]}
longitudes = {"rev":[], "rate":[]}

#Both Review Ratings
if reviews:
    for review in reviews:
        latitudes['rev'].append(float(review['latitude']))
        longitudes['rev'].append(float(review['longitude']))

if ratings:        
    for rating in ratings:
        latitudes['rate'].append(float(rating['latitude']))
        longitudes['rate'].append(float(rating['longitude']))

if latitudes:
    basemap = gmplot.GoogleMapPlotter(latitudes['rev'][0], longitudes['rev'][0], 12)  
    basemap.scatter(latitudes['rev'][1:], longitudes['rev'][1:], 'blue', size = 40)
    basemap.scatter(latitudes['rate'], longitudes['rate'], 'orange', size = 40)
    basemap_file = targetUser +"/_review&rating_basemap.html"
    basemap.draw(basemap_file)
    
    heatmap = gmplot.GoogleMapPlotter(latitudes['rev'][0], longitudes['rev'][0], 12)
    heatmap.heatmap(latitudes['rev'][1:], longitudes['rev'][1:])
    heatmap.heatmap(latitudes['rate'], longitudes['rate'])
    heatmap_file = targetUser + "/_review&rating_heatmap.html"
    heatmap.draw(heatmap_file)
    
```
